```python
# se importan todas las librerias
import pandas as pd 
import numpy as np
import kagglehub
import os

```

```python
# descargar ultima version de ventas
path = kagglehub.dataset_download("chadwambles/supermarket-sales")
print(os.listdir(path))
```

```python
# creo un dataframe con el archivo csv descargado
df =pd.read_csv(path + "/sales.csv")
```

```python
# cambiar nombres de encabezados


df = df.rename(columns={
    'sale_id':'Id_ventas',
    'branch': 'Rama', 
    'city':'ciudad',
    'customer_type':'Tipo_de_cliente', 
    'gender': 'genero', 
    'product_name':'nombre_producto',
    'product_category':'categoria_producto', 
    'unit_price':'precio_unitario', 
    'quantity':'cantidad', 
    'tax':'impuesto', 
    'total_price':'precio_total',
    'reward_points':'puntos'
})
```

```python
tipo_cliente = df['Tipo_de_cliente'].unique().tolist()
lista_genero = df['genero'].unique().tolist()
categoria_prod = df['categoria_producto'].unique().tolist()
productos = df['nombre_producto'].unique().tolist()

tipo_cliente_NEW = ['Miembro','Normal']
lista_genero_NEW = ['Masculino','Femenino']
categoria_prod_NEW = ['Cuidado personal','Papelería','Frutas','Hogar','Bebidas' ]
productos_NEW = ['Shampoo','Cuaderno','Manzana','Detergente','Jugo de naranja']
```

```python
# reemplazar valores en ingles
def reemplazar (viejos, nuevos, df,  columna):
    df[columna] = df[columna].replace(viejos,nuevos)
    
reemplazar(tipo_cliente,tipo_cliente_NEW,df,'Tipo_de_cliente')    
reemplazar(lista_genero,lista_genero_NEW,df,'genero')
reemplazar(categoria_prod,categoria_prod_NEW,df,'categoria_producto')
reemplazar(productos,productos_NEW,df,'nombre_producto')
```

```python
df.loc[df['nombre_producto'] == 'Shampoo','categoria_producto'] = 'Cuidado personal'
df.loc[df['nombre_producto'] == 'Cuaderno','categoria_producto'] = 'Papelería'
df.loc[df['nombre_producto'] == 'Manzana','categoria_producto'] = 'Frutas'
df.loc[df['nombre_producto'] == 'Detergente','categoria_producto'] = 'Hogar'
df.loc[df['nombre_producto'] == 'Jugo de naranja','categoria_producto'] = 'Bebidas'
```

```python
df.to_excel('tienda.xlsx')
```
